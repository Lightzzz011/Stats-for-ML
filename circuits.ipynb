```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"circuits.csv", encoding="latin1")
```

```python
avg_latitude = df["lat"].mean()
avg_latitude
```

```python
longitude_distribution = df["long"].describe()
longitude_distribution
```

```python
europe = ["UK", "Monaco", "Switzerland"]
median_lat_eur = df[df["country"].isin(europe)]["lat"].median()
median_lat_eur
```

```python
locality_counts = df["locality"].value_counts()
locality_counts
```

```python
country_lat_spread = df.groupby("country")["lat"].std()
country_lat_spread
```
